```python
import pandas as pd
import datetime

## Loading data properly
epoch = datetime.datetime.utcfromtimestamp(0)
date_to_epoch = lambda x: (datetime.datetime(x.year, x.month, 1, 1, 0) - epoch).total_seconds() * 1000.0
ns_data = pd.read_csv("data/delays.csv", parse_dates=["start_time"])
ns_data["edate"] = ns_data["start_time"].apply(lambda x: date_to_epoch(x.date()))
ns_data["date"] = ns_data["start_time"].apply(lambda x: x.date())
ns_data = ns_data[["ns_lines", "statistical_cause_en", "cause_group", "edate", "date"]]
ns_data = ns_data.set_index(ns_data["date"]).drop(["date", "edate"], axis = 1)
ns_data.to_csv("delays.csv")
```

```python
import pandas as pd
import numpy as np
weather = pd.read_csv("data/weather.csv", skiprows=range(15))
data = pd.DataFrame()
for col in weather.columns:
    weather = weather[weather[col].apply(lambda x: bool(str(x).strip()))]
    weather[col] = weather[col].apply(lambda x: int(str(x).strip()))
weather = weather.loc[weather["YYYYMMDD"] >= 20110101].reset_index(drop=True).drop("STN", axis=1)
weather["date"] = weather["YYYYMMDD"].apply(lambda x: datetime.datetime(int(str(x)[0:4]), int(str(x)[4:6]), int(str(x)[6:8])))
weather = weather.drop("YYYYMMDD", axis=1).set_index("date")
weather
```

```python
import numpy as np

avg_weather = weather.groupby("date").aggregate(lambda x: round(np.average(x)/10.0))
avg_weather["min_sight"] = avg_weather["min_sight"]*10
avg_weather["max_sight"] = avg_weather["max_sight"]*10
avg_weather["humidity"] = avg_weather["humidity"]*10
avg_weather["min_humidity"] = avg_weather["min_humidity"]*10
avg_weather.to_csv("weather.csv")
avg_weather
```
